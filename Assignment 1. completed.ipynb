```python
# 4. find Marks Average/Aggregate Marks:

marks=[]
sci = float(input("Enter your sci marks:"))
math =float(input("Enter your math marks:"))
ss = float(input("Enter your ss marks:"))
eco = float(input("Enter your eco marks:"))
eng=float(input("Enter your eng marks:"))
if sci > 100 or sci <0 or math > 100 or math <0 or ss > 100 or ss <0 or eco > 100 or eco <0 or eng > 100 or eng <0:
    print("Invalid Marks:")
else:
    Aggregate = sci+math+ss+eco+eng
    Avg = Aggregate/5
    print("Aggregate marks:",Aggregate)
    print("Avg marks:",Avg)
```
